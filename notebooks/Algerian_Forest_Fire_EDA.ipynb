## Data Set Information:

The dataset includes 244 instances that regroup a data of two regions of Algeria,namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012. The dataset includes 11 attribues and 1 output attribue (class) The 244 instances have been classified into fire(138 classes) and not fire (106 classes) classes.

Attribute Information:

Date : (DD/MM/YYYY) Day, month ('june' to 'september'), year (2012) Weather data observations

Temp : temperature noon (temperature max) in Celsius degrees: 22 to 42

RH : Relative Humidity in %: 21 to 90

Ws :Wind speed in km/h: 6 to 29

Rain: total day in mm: 0 to 16.8 FWI Components

Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5

Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9

Drought Code (DC) index from the FWI system: 7 to 220.4

Initial Spread Index (ISI) index from the FWI system: 0 to 18.5

Buildup Index (BUI) index from the FWI system: 1.1 to 68

Fire Weather Index (FWI) Index: 0 to 31.1

Classes: two classes, namely Fire and not Fire

## DATA CLEANING

```python
# check for missing values
dataset.isnull().sum()
```

```python
dataset.isnull().any(axis=1)
```

```python
dataset[dataset.isnull().any(axis=1)] # retrieve those records which r true
```

The dataset is converted into two sets based on Region from 122th index, we can make a new column based on the Region

1 : "Bejaia Region Dataset"

2 : "Sidi-Bel Abbes Region Dataset"

Add new column with region

```python
df=dataset
df.loc[:122,"Region"]=0 # for bejala region
df.loc[122:,"Region"]=1 # for Sidi bel Abbes
```

```python
df.head(2)
```

```python
df[["Region"]]=df[["Region"]].astype(int) # convert float to int
```

```python
df.info()
```

```python
# Handle missing values
df.isnull().sum()
```

```python
# Remove null values
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.tail()
```

```python
df.iloc[[122]] # good
```

```python
df.columns
```

```python
# Fix spaces and column names
df.columns=df.columns.str.strip()
```

```python
df.columns
```

```python
# Change the required columns as integer data type
df[["month","day","year","Temperature","RH","Ws"]]=df[["month","day","year","Temperature","RH","Ws"]].astype(int)
```

```python
df.info()
```

```python
### Changing the other columns to float data type
objects = [features for features in df.columns if df[features].dtypes=="O"]
print(objects)
```

```python
for i in objects:
  if i!="Classes":
    df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
# save this new cleaned dataset
df.to_csv("Algerian_forest_fires_CLEANED_dataset.csv",index=False)
```
